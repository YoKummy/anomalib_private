```python
# Install anomalib if not already installed
# %pip install anomalib


import torch

from anomalib.data import ImageItem
from anomalib.engine import Engine
from anomalib.models import Patchcore

# Anomalib imports
from anomalib.visualization import ImageVisualizer
from anomalib.visualization.image.item_visualizer import visualize_image_item

print("✅ All imports successful!")
```

```python
# Create visualizer with default settings
visualizer = ImageVisualizer()

# Create model with visualizer
model = Patchcore(
    visualizer=visualizer,  # Pass visualizer to the model
)

# Create engine
engine = Engine(logger=False, enable_checkpointing=False)

print("✅ Model and visualizer created successfully!")
print(f"Model: {type(model).__name__}")
print(f"Visualizer: {type(visualizer).__name__}")
```

```python
# Create sample data for demonstration
def create_sample_data() -> tuple[torch.Tensor, torch.Tensor, torch.Tensor, torch.Tensor]:
    """Create sample image and mask data for demonstration."""
    # Create a sample image (RGB)
    image = torch.randint(0, 255, (3, 256, 256), dtype=torch.uint8)

    # Create a sample ground truth mask
    gt_mask = torch.zeros((256, 256), dtype=torch.bool)
    gt_mask[100:150, 100:150] = True  # Add anomalous region

    # Create a sample anomaly map
    anomaly_map = torch.rand(256, 256)
    anomaly_map[100:150, 100:150] *= 2  # Higher values in anomalous region

    # Create a sample prediction mask
    pred_mask = torch.zeros((256, 256), dtype=torch.bool)
    pred_mask[95:155, 95:155] = True  # Slightly different from GT

    return image, gt_mask, anomaly_map, pred_mask


# Generate sample data
sample_image, sample_gt_mask, sample_anomaly_map, sample_pred_mask = create_sample_data()

print("✅ Sample data created:")
print(f"  Image shape: {sample_image.shape}")
print(f"  GT mask shape: {sample_gt_mask.shape}")
print(f"  Anomaly map shape: {sample_anomaly_map.shape}")
print(f"  Pred mask shape: {sample_pred_mask.shape}")
```

```python
# Create an ImageItem
item = ImageItem(
    image_path="sample_image.png",
    mask_path="sample_mask.png",
    image=sample_image,
    gt_mask=sample_gt_mask,
    anomaly_map=sample_anomaly_map,
    pred_mask=sample_pred_mask,
)

# Generate visualization
vis_result = visualize_image_item(item, fields=["image", "gt_mask"])

print("✅ Visualization created successfully!")
print(f"Visualization type: {type(vis_result)}")
```
